```python
import numpy as np
import pandas as pd
from scipy.stats import mstats
import matplotlib.pyplot as plt
import seaborn as sns
import pystan
%matplotlib inline
plt.style.use("ggplot")
```

```python
data = pd.read_csv('./input/data-conc-2.txt')
data.head()
```

```python
sns.pairplot(data)
```

```python
data_melt = data.melt(id_vars=['PersonID'])
data_melt['variable'] = data_melt['variable'].replace({'Time1':1, 'Time2':2, 'Time4':4, 'Time8':8, 'Time12':12, 'Time24':24})
```

```python
data_melt.head()
```

```python
sns.lineplot(data=data_melt, x='variable', y='value', hue='PersonID', sort=False, palette='muted')
```

```python
stan_model = """

data {
  int N;
  int T;
  real Time[T];
  real Y[N, T];
  int T_new;
  real Time_new[T_new];
}


parameters {
  real a0;
  real b0;
  real log_a[N];
  real log_b[N];
  real<lower=0> s_a;
  real<lower=0> s_b;
  real<lower=0> s_Y;
}

transformed parameters {
  real<lower=0, upper=100> a[N];
  real<lower=0, upper=5> b[N];
  for (n in 1:N){
    a[n] = exp(log_a[n]);
    b[n] = exp(log_b[n]);
  }
}

model {
  for (n in 1:N){
    log_a[n] ~ normal(a0, s_a);
    log_b[n] ~ normal(a0, s_b);    
  }
  
  for (n in 1:N){
    for (t in 1:T){
      Y[n, t] ~ normal(a[n] * (1 - exp(-b[n] * Time[t])), s_Y);
    }
  }
}

generated quantities {
  real y_new[N, T_new];
  real y_pred[N, T];  
  
  for (n in 1:N){
    for (t in 1:T_new){
      y_new[n, t] = normal_rng(a[n] * (1 - exp(-b[n] * Time_new[t])), s_Y);
    }
  }
  
  for (n in 1:N){
    for (t in 1:T){
      y_pred[n, t] = normal_rng(a[n] * (1 - exp(-b[n] * Time[t])), s_Y);
    }
  }  
  
}
"""
```

```python
sm = pystan.StanModel(model_code=stan_model)
```

```python
stan_data = {}
stan_data['N'] = data.shape[0]
stan_data['Time'] = np.sort(data_melt['variable'].unique())
stan_data['T'] = len(stan_data['Time'])
stan_data['Y'] = data.drop(['PersonID'], axis=1)
stan_data['Time_new'] = [60]
stan_data['T_new'] = len(stan_data['Time_new'])
```

```python
fit = sm.sampling(stan_data, seed=1234)
```

```python
fit
```

```python
fig = fit.plot()
```

```python
sample_a = fit.extract('a')['a']
sample_b = fit.extract('b')['b']
sample_y_pred = fit.extract('y_pred')['y_pred']
```

```python
#sns.lineplot(data=data_melt[data_melt['PersonID']==1], x='variable', y='value', sort=False, palette='muted')
#sns.lineplot(data=data_melt, x='variable', y='value', hue='PersonID', sort=False, palette='muted')
cp = sns.color_palette('muted', n_colors=16)
fig, axes = plt.subplots(4, 4, figsize=(10, 10))

for pid in sorted(data['PersonID'].unique()):
#for pid in [1]:
    data_melt2 = data_melt[data_melt['PersonID']==pid]
    sample_y_pred_pid = sample_y_pred[:,pid-1,:]
    sample_a_pid = sample_a[:,pid-1]
    sample_b_pid = sample_b[:,pid-1]
    
    lower80, upper80 = mstats.mquantiles(sample_y_pred_pid, [0.1, 0.9], axis=0)
    sample_y_pred_mean = np.mean(sample_y_pred_pid, axis=0)
    sample_a_mean = np.mean(sample_a_pid, axis=0)
    sample_b_mean = np.mean(sample_b_pid, axis=0)    
    
    plot_r= int((pid - 1) / 4)
    plot_c = (pid -1 ) % 4
    ax = axes[plot_r, plot_c]
    
    sns.lineplot(data=data_melt2, x='variable', y='value', sort=False, color=cp[pid-1], ax=ax)
    lower_dist = sample_y_pred_mean - lower80
    upper_dist = upper80 -  sample_y_pred_mean
    ax.errorbar(data_melt2['variable'], sample_y_pred_mean, [lower_dist, upper_dist], ls='', alpha=0.6, color='gray')

    x = np.arange(min(data_melt2['variable']), max(data_melt2['variable']+1), 0.1)
    y = sample_a_mean * (1 - np.exp(-sample_b_mean * x))
    sns.lineplot(x, y, color=cp[pid-1], ax=ax)
```
