```python
import numpy as np
import pandas as pd
from scipy.stats import mstats
import matplotlib.pyplot as plt
import seaborn as sns
import pystan
%matplotlib inline
plt.style.use("ggplot")
```

```python
data = pd.read_csv('./input/data-salary-2.txt')
data.head()
```

```python
stan_model3 = """
data {
  int N;
  int K;
  real X[N];
  real Y[N];
  int<lower=1, upper=K> KID[N];
}

parameters {
  real a0;
  real b0;
  real a[K];
  real b[K];
  real<lower=0> s_a;
  real<lower=0> s_b;
  real<lower=0> s_Y;
}

model {
  for (k in 1:K){
    a[k] ~ normal(a0, s_a);
    b[k] ~ normal(b0, s_b);
  }

  for (n in 1:N){
    Y[n] ~ normal(a[KID[n]] + b[KID[n]] * X[n], s_Y);
    }
}

generated quantities {
  real y_pred[N];
  for (n in 1:N){
    y_pred[n] = normal_rng(a[KID[n]] + b[KID[n]] * X[n], s_Y);
  } 
}
"""
```

```python
sm3 = pystan.StanModel(model_code=stan_model3)
```

```python
stan_data = {}
stan_data['N'] = data.shape[0]
stan_data['K'] = len(data['KID'].unique())
stan_data['X'] = data['X']
stan_data['Y'] = data['Y']
stan_data['KID'] = data['KID']
```

```python
fit3 = sm3.sampling(stan_data)
```

```python
fit3
```

```python
fig = fit3.plot()
```

```python
stan_model3b = """
data {
  int N;
  int K;
  vector[N] X;
  vector[N] Y;
  int<lower=1, upper=K> KID[N];
}

parameters {
  real a0;
  real b0;
  vector[K] a;
  vector[K] b;
  real<lower=0> s_a;
  real<lower=0> s_b;
  real<lower=0> s_Y;
}

model {
  a ~ normal(a0, s_a);
  b ~ normal(b0, s_b);
  Y ~ normal(a[KID] + b[KID] .* X, s_Y);
}

generated quantities {
  real y_pred[N];
  for (n in 1:N){
    y_pred[n] = normal_rng(a[KID[n]] + b[KID[n]] * X[n], s_Y);
  } 
}
"""
```

```python
sm3b = pystan.StanModel(model_code=stan_model3b)
```

```python
fit3b = sm3b.sampling(stan_data)
```

```python
fit3b
```

```python
fig = fit3b.plot()
```

```python
fit3b = sm3b.sampling(stan_data, iter=4000, warmup=2000)
```

```python
fit3b
```

```python
fig = fit3b.plot()
```
