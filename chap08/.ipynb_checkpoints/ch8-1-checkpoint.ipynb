```python
import numpy as np
import pandas as pd
from scipy.stats import mstats
import matplotlib.pyplot as plt
import seaborn as sns
import pystan
%matplotlib inline
plt.style.use("ggplot")
```

```python
data = pd.read_csv('./input/data-salary-2.txt')
data.head()
```

```python
sns.pairplot(data)
```

```python
sns.pairplot(data, hue='KID')
```

```python
sns.scatterplot(data=data, x='X', y='Y', hue='KID', palette='muted')
```

```python
stan_model1 = """
data {
  int N;
  real X[N];
  real Y[N];
}

parameters {
  real a;
  real b;
  real<lower=0> s_Y;
}

model {
  for (n in 1:N){
    Y[n] ~ normal(a + b * X[n], s_Y);
    }
}

generated quantities {
  real y_pred[N];
  for (n in 1:N){
    y_pred[n] = normal_rng(a + b * X[n], s_Y);
  } 
}
"""
```

```python
sm1 = pystan.StanModel(model_code=stan_model1)
```

```python
stan_data = {}
stan_data['N'] = data.shape[0]
stan_data['X'] = data['X']
stan_data['Y'] = data['Y']
```

```python
fit1 = sm1.sampling(stan_data)
```

```python
fit1
```

```python
fig = fit1.plot()
```

```python
sample_a_1 = fit1.extract('a')['a']
sample_b_1 = fit1.extract('b')['b']
sample_y_pred_1 = fit1.extract('y_pred')['y_pred']
```

```python
lower80, upper80 = mstats.mquantiles(sample_y_pred_1, [0.1, 0.9], axis=0)
sample_y_pred_mean_1 = np.mean(sample_y_pred_1, axis=0)

sns.scatterplot(data['Y'], sample_y_pred_mean_1, data['KID'], palette='muted')
lower_dist = sample_y_pred_mean_1 - lower80
upper_dist = upper80 -  sample_y_pred_mean_1
plt.errorbar(data['Y'], sample_y_pred_mean_1, [lower_dist, upper_dist], ls='', alpha=0.6, color='gray')

x = np.arange(min(data['Y']), max(data['Y']))
y = x
plt.plot(x, y, color='black')
```

```python
sns.scatterplot(data['X'], data['Y'], hue=data['KID'], palette='muted')
lower80, upper80 = mstats.mquantiles(sample_y_pred_1, [0.1, 0.9], axis=0)
sample_y_pred_mean_1 = np.mean(sample_y_pred_1, axis=0)
sample_a_mean_1 = np.mean(sample_a_1, axis=0)
sample_b_mean_1 = np.mean(sample_b_1, axis=0)    
    

lower_dist = sample_y_pred_mean_1 - lower80
upper_dist = upper80 -  sample_y_pred_mean_1
plt.errorbar(data['X'], sample_y_pred_mean_1, [lower_dist, upper_dist], ls='', alpha=0.6, color='gray')

x = np.arange(min(data['X']), max(data['X']+1))
y = sample_a_mean_1 + sample_b_mean_1 * x
plt.plot(x, y, color='black')
```

```python
stan_model2 = """
data {
  int N;
  int K;
  real X[N];
  real Y[N];
  int<lower=1, upper=K> KID[N];
}

parameters {
  real a[K];
  real b[K];
  real<lower=0> s_Y;
}

model {
  for (n in 1:N){
    Y[n] ~ normal(a[KID[n]] + b[KID[n]] * X[n], s_Y);
    }
}

generated quantities {
  real y_pred[N];
  for (n in 1:N){
    y_pred[n] = normal_rng(a[KID[n]] + b[KID[n]] * X[n], s_Y);
  } 
}
"""
```

```python
sm2 = pystan.StanModel(model_code=stan_model2)
```

```python
stan_data['K'] = len(data['KID'].unique())
stan_data['KID'] = data['KID']
```

```python
fit2 = sm2.sampling(stan_data)
```

```python
fit2
```

```python
fig = fit2.plot()
```

```python
sample_a_2 = fit2.extract('a')['a']
sample_b_2 = fit2.extract('b')['b']
sample_y_pred_2 = fit2.extract('y_pred')['y_pred']
```

```python
lower80_2, upper80_2 = mstats.mquantiles(sample_y_pred_2, [0.1, 0.9], axis=0)
sample_y_pred_mean_2 = np.mean(sample_y_pred_2, axis=0)

sns.scatterplot(data['Y'], sample_y_pred_mean_2, data['KID'], palette='muted')
lower_dist_2 = sample_y_pred_mean_2 - lower80_2
upper_dist_2 = upper80_2 -  sample_y_pred_mean_2
plt.errorbar(data['Y'], sample_y_pred_mean_2, [lower_dist_2, upper_dist_2], ls='', alpha=0.6, color='gray')

x = np.arange(min(data['Y']), max(data['Y']))
y = x
plt.plot(x, y, color='black')
```

```python
sns.scatterplot(data['X'], data['Y'], hue=data['KID'], palette='muted')
cp = sns.color_palette('muted')

for kid in sorted(data['KID'].unique()):
    data2 = data[data['KID']==kid]
    sample_y_pred_2_kid = sample_y_pred_2[:,(data['KID']==kid)]
    sample_a_2_kid = sample_a_2[:,kid-1]
    sample_b_2_kid = sample_b_2[:,kid-1]
    
    lower80, upper80 = mstats.mquantiles(sample_y_pred_2_kid, [0.1, 0.9], axis=0)
    sample_y_pred_mean_2 = np.mean(sample_y_pred_2_kid, axis=0)
    sample_a_mean_2 = np.mean(sample_a_2_kid, axis=0)
    sample_b_mean_2 = np.mean(sample_b_2_kid, axis=0)    
    
    sns.scatterplot(data2['X'], data2['Y'], color=cp[kid-1])
    lower_dist = sample_y_pred_mean_2 - lower80
    upper_dist = upper80 -  sample_y_pred_mean_2
    plt.errorbar(data2['X'], sample_y_pred_mean_2, [lower_dist, upper_dist], ls='', alpha=0.6, color='gray')

    x = np.arange(min(data2['X']), max(data2['X']+1))
    y = sample_a_mean_2 + sample_b_mean_2 * x
    plt.plot(x, y, color=cp[kid-1])
```

```python
stan_model3 = """
data {
  int N;
  int K;
  real X[N];
  real Y[N];
  int<lower=1, upper=K> KID[N];
}

parameters {
  real a0;
  real b0;
  real ak[K];
  real bk[K];
  real<lower=0> s_a;
  real<lower=0> s_b;
  real<lower=0> s_Y;
}

transformed parameters {
  real a[K];
  real b[K];
  for (k in 1:K){
    a[k] = a0 + ak[k];
    b[k] = b0 + bk[k];
  }
}

model {
  for (k in 1:K){
    ak[k] ~ normal(0, s_a);
    bk[k] ~ normal(0, s_b);
  }

  for (n in 1:N){
    Y[n] ~ normal(a[KID[n]] + b[KID[n]] * X[n], s_Y);
    }
}

generated quantities {
  real y_pred[N];
  for (n in 1:N){
    y_pred[n] = normal_rng(a[KID[n]] + b[KID[n]] * X[n], s_Y);
  } 
}
"""
```

```python
sm3 = pystan.StanModel(model_code=stan_model3)
```

```python
fit3 = sm3.sampling(stan_data)
```

```python
fit3
```

```python
fig = fit3.plot()
```

```python
sample_a_3 = fit3.extract('a')['a']
sample_b_3 = fit3.extract('b')['b']
sample_y_pred_3 = fit3.extract('y_pred')['y_pred']
```

```python
lower80, upper80 = mstats.mquantiles(sample_y_pred_3, [0.1, 0.9], axis=0)
sample_y_pred_mean = np.mean(sample_y_pred_3, axis=0)

sns.scatterplot(data['Y'], sample_y_pred_mean, data['KID'], palette='muted')
lower_dist = sample_y_pred_mean - lower80
upper_dist = upper80 -  sample_y_pred_mean
plt.errorbar(data['Y'], sample_y_pred_mean, [lower_dist, upper_dist], ls='', alpha=0.6, color='gray')

x = np.arange(min(data['Y']), max(data['Y']))
y = x
plt.plot(x, y, color='black')
```

```python
sns.scatterplot(data['X'], data['Y'], hue=data['KID'], palette='muted')
cp = sns.color_palette('muted')
for kid in sorted(data['KID'].unique()):
    data2 = data[data['KID']==kid]
    sample_y_pred_3_kid = sample_y_pred_3[:,(data['KID']==kid)]
    sample_a_3_kid = sample_a_3[:,kid-1]
    sample_b_3_kid = sample_b_3[:,kid-1]
    
    lower80, upper80 = mstats.mquantiles(sample_y_pred_3_kid, [0.1, 0.9], axis=0)
    sample_y_pred_mean_3 = np.mean(sample_y_pred_3_kid, axis=0)
    sample_a_mean_3 = np.mean(sample_a_3_kid, axis=0)
    sample_b_mean_3 = np.mean(sample_b_3_kid, axis=0)    
    
    sns.scatterplot(data2['X'], data2['Y'], color=cp[kid-1])
    lower_dist = sample_y_pred_mean_3 - lower80
    upper_dist = upper80 -  sample_y_pred_mean_3
    plt.errorbar(data2['X'], sample_y_pred_mean_3, [lower_dist, upper_dist], ls='', alpha=0.6, color='gray')

    x = np.arange(min(data2['X']), max(data2['X']+1))
    y = sample_a_mean_3 + sample_b_mean_3 * (x)
    plt.plot(x, y, color=cp[kid-1])
```
