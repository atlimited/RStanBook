```python
import numpy as np
import pandas as pd
from scipy.stats import mstats
import matplotlib.pyplot as plt
import seaborn as sns
import pystan
%matplotlib inline
plt.style.use("ggplot")
```

```python
data = pd.read_csv('./input/data-ZIP.txt')
data.head()
```

```python
sns.pairplot(data)
```

```python
data['Age'] /= 10
```

```python
stan_model = """

functions {
  real ZIP_lpmf(int Y, real q, real lambda){
    if (Y == 0){
      return log_sum_exp(
        bernoulli_lpmf(0 | q),
        bernoulli_lpmf(1 | q) + poisson_log_lpmf(0 | lambda)
      );
    } else {
      return bernoulli_lpmf(1 | q) + poisson_log_lpmf(Y | lambda);
    }
  }
}

data {
  int N;
  int D;
  int<lower=0> Y[N];
  matrix[N, D] X;
}

parameters {
  vector[D] b[2];
}


transformed parameters {
  vector[N] q_x;
  vector[N] q;
  vector[N] lambda;
  
  q_x = X * b[1];
  lambda = X * b[2];
  for (n in 1:N){
    q[n] = inv_logit(q_x[n]);
  }  
}

model {
  for (n in 1:N){
    Y[n] ~ ZIP(q[n], lambda[n]);
  }
}

"""
```

```python
sm = pystan.StanModel(model_code=stan_model)
```

```python
stan_data = {}
stan_data['N'] = data.shape[0]
stan_data['D'] = data.shape[1]
stan_data['Y'] = data['Y']
intercept = pd.DataFrame({'intercept': np.ones(data.shape[0])})
stan_data['X'] = pd.concat([intercept, data['Sex'], data['Sake'], data['Age']], axis=1)
```

```python
fit = sm.sampling(stan_data)
```

```python
fit
```

```python
fig = fit.plot()
```

```python
sample_b1 = fit.extract('b')['b'][:,0,:]
sample_b2 = fit.extract('b')['b'][:,1,:]
```

```python
vio_df = pd.concat([
    pd.DataFrame({'param':'intersept', 'value':sample_b1[:,0]}),
    pd.DataFrame({'param':'Sex', 'value':sample_b1[:,1]}),    
    pd.DataFrame({'param':'Sake', 'value':sample_b1[:,2]}),
    pd.DataFrame({'param':'Age', 'value':sample_b1[:,3]})
])

sns.violinplot(data=vio_df, x='param', y='value')
```

```python
vio_df = pd.concat([
    pd.DataFrame({'param':'intersept', 'value':sample_b2[:,0]}),
    pd.DataFrame({'param':'Sex', 'value':sample_b2[:,1]}),    
    pd.DataFrame({'param':'Sake', 'value':sample_b2[:,2]}),
    pd.DataFrame({'param':'Age', 'value':sample_b2[:,3]})
])

sns.violinplot(data=vio_df, x='param', y='value')
```
