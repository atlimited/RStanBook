```python
import numpy as np
import pandas as pd
from scipy.stats import mstats
import matplotlib.pyplot as plt
import seaborn as sns
import pystan
%matplotlib inline
plt.style.use("ggplot")
```

```python
data = pd.read_csv('./input/data-mix1.txt')
data.head()
```

```python
sns.distplot(data)
```

```python
stan_model = """

data {
  int N;
  vector[N] Y;
}

parameters {
  real<lower=0, upper=1> a;
  ordered[2] mu;
  vector<lower=0, upper=10>[2] sigma; // upper=10
}


model {
  for (n in 1:N){
    target += log_sum_exp(
      log(a) + normal_lpdf(Y[n] | mu[1], sigma[1]),
      log1m(a) + normal_lpdf(Y[n] | mu[2], sigma[2])
    );
  }
}

"""
```

```python
sm = pystan.StanModel(model_code=stan_model)
```

```python
stan_data = {}
stan_data['N'] = data.shape[0]
stan_data['Y'] = data['Y']
```

```python
fit = sm.sampling(stan_data, seed=1234, iter=3000)
```
