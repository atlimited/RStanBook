```python
import numpy as np
import pandas as pd
from scipy.stats import mstats
import matplotlib.pyplot as plt
import seaborn as sns
import pystan
%matplotlib inline
plt.style.use("ggplot")
```

```python
data = pd.read_csv('./input/data-kubo11a.txt')
data.head()
```

```python
sns.lineplot(data=data, x=data.index, y='Y')
```

```python
stan_model = """

data {
  int I;
  int Y[I];
}

parameters {
  vector[I] r;
  real<lower=0> s_r;
}

model {
  target += normal_lpdf(r[3:I] | 2 * r[2:(I-1)] - r[1:(I-2)], s_r);
  Y ~ poisson_log(r);
}

generated quantities {
  vector[I] Y_mean;
  Y_mean = exp(r);
}
"""
```

```python
sm = pystan.StanModel(model_code=stan_model)
```

```python
stan_data = {}
stan_data['I'] = data.shape[0]
stan_data['Y'] = data['Y']
```

```python
fit = sm.sampling(stan_data)
```

```python
fit
```

```python
fig = fit.plot()
```

```python
sample_y_mean = fit.extract('Y_mean')['Y_mean']
lower80, upper80 = mstats.mquantiles(sample_y_mean, [0.1, 0.9], axis=0)
plt.scatter(data.index, data['Y'])
plt.fill_between(data.index, lower80, upper80, alpha=0.6, color='darkgray')
sns.lineplot(data=data, x=data.index, y='Y')

x = data.index
y = np.mean(sample_y_mean, axis=0)
plt.plot(x, y, color='black')
```
