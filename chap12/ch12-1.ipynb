```python
import numpy as np
import pandas as pd
from scipy.stats import mstats
import matplotlib.pyplot as plt
import seaborn as sns
import pystan
%matplotlib inline
plt.style.use("ggplot")
```

```python
data = pd.read_csv('./input/data-ss1.txt')
data.head()
```

```python
sns.lineplot(data=data, x='X', y='Y')
```

```python
stan_model = """

data {
  int T;
  int T_pred;
  vector[T] Y;
}

parameters {
  vector[T] mu;
  real<lower=0> s_mu;
  real<lower=0> s_Y;
}

model {
  mu[2:T] ~ normal(mu[1:(T-1)], s_mu);
  Y ~ normal(mu, s_Y);
}

generated quantities {
  vector[T+T_pred] mu_all;
  vector[T_pred] y_pred;
  mu_all[1:T] = mu;
  for (t in 1:T_pred){
    mu_all[T+t] = normal_rng(mu_all[T+t-1], s_mu);
    y_pred[t] = normal_rng(mu_all[T+t], s_Y);  
  }
}

"""
```

```python
sm = pystan.StanModel(model_code=stan_model)
```

```python
stan_data = {}
stan_data['T'] = data.shape[0]
stan_data['T_pred'] = 3
stan_data['Y'] = data['Y']
```

```python
fit = sm.sampling(stan_data)
```

```python
fit
```

```python
fig = fit.plot()
```

```python
sample_mu = fit.extract('mu_all')['mu_all']
```

```python
lower80, upper80 = mstats.mquantiles(sample_mu, [0.1, 0.9], axis=0)
x = np.arange(data.shape[0] + 3) + 1
y = np.mean(sample_mu, axis=0) 

plt.plot(x, y, color='black')
plt.fill_between(x, lower80, upper80, alpha=0.6, color='darkgray')

plt.scatter(data['X'], data['Y'])
```
